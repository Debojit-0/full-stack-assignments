```python
#2
from collections import Counter

def is_valid_string(s):
    frequencies = Counter(s)
    print(frequencies)
    print(frequencies.values())
    values = set(frequencies.values())
    print(values)
    print(len(values))

    if len(values) == 1:
        return "YES"

    if len(values) == 2:
        min_freq = min(values)
        print(min_freq)
        max_freq = max(values)
        if min_freq == 1 and list(frequencies.values()).count(min_freq) == 1:
            return "YES"
        if max_freq - min_freq == 1 and list(frequencies.values()).count(max_freq) == 1:
            return "YES"

    return "NO"

# Example usage
print(is_valid_string("aabbcccc"))  # Output: YES
#print(is_valid_string("aabbcc"))  # Output: YES
#print(is_valid_string("abc"))  # Output: YES
#print(is_valid_string("abcc"))  # Output: NO


```
